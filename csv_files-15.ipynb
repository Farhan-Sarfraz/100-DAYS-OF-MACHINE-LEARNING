# converts 
here we convert big names or data into thier short forms 
means we can convert royal challenges bangloru into RCB


```python
pd.read_csv("IPL Matches.csv")
```

```python
def rename(name):
    if name == "Mumbai Indians":
        return 'MI'
    elif name == "Royal Challengers Bangalore":
        return "RCB"
    elif name == "Sunrisers Hyderabad":
        return 'SRH'
    elif name == "Delhi Capitals":
        return 'DC'
    elif name == 'Kolkata Knight Riders':
        return 'kkr'
    elif name == 'Deccan Chargers':
        return 'DCC'
    elif name == 'Rajasthan Royals':
        return 'RR'
    elif name == 'Delhi Daredevils':
        return 'DDL'
    else:
        return name
        


    
```

```python
pd.read_csv("IPL Matches.csv" , converters = {'team1' : rename,'team2' : rename})
```
