# parameters practice 



below are the few important parameters

```python
pd.read_sql_query("select * From city  " , conn )
```

```python
pd.read_sql_query("select * from city " , conn , index_col = 'ID' )
```

```python
pd.read_sql_query("select * from city " , conn , parse_dates = ['order_dates'])
```

```python
chunks = pd.read_sql_query("select * from city " , conn , chunksize = 1000)
```

```python
for chunk in chunks:
    print(chunk.shape)
```

```python
chunks = pd.read_sql_query('select * from city ' , conn , chunksize = 500 )
```

```python
for chunk in chunks:
    print('chunks shape is  ' , chunk.shape)
```

```python
pd.read_sql_query('select ID , Population from city ' , conn , dtype={ 'ID' : int , 'Population' : float })
```

```python
pd.read_sql_query('select ID , Population from city ' , conn , dtype={ 'ID' : int , 'Population' : int })
```
