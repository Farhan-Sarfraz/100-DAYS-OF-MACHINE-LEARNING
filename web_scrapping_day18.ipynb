```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
quotes = soup.find_all("span", class_="text")
```

```python
quotes
```

```python
len(quotes)
```

```python
quote_list = []
for i in quotes:
    print(i.find_all('span'))
```

```python
quote_list = []
for i in quotes:
    quote_list.append(i.text)
```

```python
quote_list
```

```python
quote_list = []
for i in quotes:
    quote_list.append(i.text.strip())
```

```python
quote_list
```

```python
df = pd.DataFrame(quote_list, columns= ['quote'])
df.head()
```
